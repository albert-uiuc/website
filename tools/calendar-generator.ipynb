```python
start = datetime.datetime.strptime(start, '%Y-%m-%d')
print("start: " + str(start.date()) + " (" + start.strftime("%a") + ")")
end = datetime.datetime.strptime(end, '%Y-%m-%d')
print("end: " + str(end.date()) + " (" + end.strftime("%a") + ")")

while start.weekday() != 0:
    start += datetime.timedelta(-1)
print("extended start: " + str(start.date()) + " (" + start.strftime("%a") + ")")
while end.weekday() != 4:
    end += datetime.timedelta(1)
print("extended end: " + str(end.date()) + " (" + end.strftime("%a") + ")")
```

```python
dtrange = pd.date_range(start = start, end = end, freq = 'B')

cal = pd.DataFrame(dtrange)


cal.insert(1, 'month', dtrange.strftime("%B"))
cal.insert(2, 'day', dtrange.strftime("%a"))

ct = 0
week = []
for i in range(0,cal.shape[0]):
    week.append(int(i/5)+1)
#     print(i)
#     print(int(i/5)+1)
#     print(week)
#     if ct >= 5:
#         print('big')
#         ct = 0
#     ct += 1
#     print(ct)
#     if ct == 1 or ct == 3 or ct == 5:
#         week.append(int(i/5)+1)
#     else:
#         pass
#     print('\n')


cal.insert(3, 'week', pd.Series(week))

l = []
for week in range(1, cal['week'].nunique()+1):
    l.append(cal[cal['week'] == week]['month'].mode()[0])

block = []
for week in cal['week']:
    block.append(l[week-1])
# df_filtered = df[df['Age'] >= 25] 
# df_new = df.drop(df[(df['col_1'] == 1.0) & (df['col_2'] == 0.0)].index)

cal.insert(4, 'block', pd.Series(block))
# df = cal[cal['day'] == 'Mon' ]
cal = cal.drop(cal[(cal['day'] == "Tue")].index)
cal = cal.drop(cal[(cal['day'] == "Thu")].index)
cal.head(10)
# cal['day'].head()
# cal.head()
# def print_full(x):
#     pd.set_option('display.max_rows', len(x))
#     print(x)
#     pd.reset_option('display.max_rows')

# print(cal.shape)
# print_full(cal)
```

```python
doc = '<div class="table-responsive">\n'
head = ('<tr>\n'+
#         '  <th>Week</th>\n'+
        '  <th>Mon</th>\n'+
#         '  <th>Tues</th>\n'+
        '  <th>Wed</th>\n'+
#         '  <th>Thurs</th>\n'+
        '  <th>Fri</th>\n'+
        '</tr>')

for block in cal['block'].value_counts()[cal['block'].unique()].index.tolist():
#     print('<h1>' + block + '</h1>')
    doc += '<h3 style="text-align: center;">' + block + '</h3>'
    doc += '\n<div class="heading-underline"></div>'
    doc +=  '\n<table class="table table-bordered">'
#     print(head)
    doc += '\n' + head + '\n'
    sub1 = cal[cal['block'] == block]
    for week in sub1['week'].value_counts()[sub1['week'].unique()].index.tolist():
        sub2 = sub1[sub1['week'] == week]
        date = ''
        content = ''
        date += '<tr>\n'
#         <td style="line-height:1px;margin:0;"></td>'
        content += '\n<tr>'
#         + '\n  <td>' + str(week_topics.get(week)) + '</td>'
        for day in sub2[0]:
            events = []
            date += '\n  <td><p class="date">' + str(day.strftime("%d")) + '</p></td>'
            if (dateContent.get(str(day.date())) != None):
                events.append(dateContent.get(str(day.date())))

            if len(events) > 1:
                content += '\n  <td>'
                n=0
                for event in events:
                    content += event
                content += '</td>'
            elif len(events) == 1:
                content += '\n  <td>' + events[0] + '</td>'
            elif len(events) == 0:
                content += '\n  <td></td>'
        content += '\n</tr>'
        date += '\n</tr>'
        doc += date + content
        doc += '\n'
#         print(date, content)
#     print('\n')
    doc += '</table>\n\n'
doc += '</div>'
# print(doc)
```
